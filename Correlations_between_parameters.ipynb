```python
import requests
import json
import pandas as pd
import time
import numpy as np
from scipy.spatial import Delaunay
import matplotlib.pyplot as plt
import matplotlib.tri as mtri
from pandas.plotting import scatter_matrix
%matplotlib inline
plt.rcParams['figure.figsize'] = (12.0, 10.0)
```

```python
def extract_data(data, device_num, start_date, end_date, parameters):
    column_names = [ f'{device_num}_{parameter}' for parameter in parameters ]
    columns_to_select = ['UTC time'] + column_names
    data_for_device = data[columns_to_select]
    data_for_device = data_for_device[data_for_device['UTC time'] >= start_date]
    data_for_device = data_for_device[data_for_device['UTC time'] <= end_date]
    return data_for_device
```

```python
locations = pd.read_csv("data/location_const.csv")
locations = locations[['id', 'latitude', 'longitude', 'city', 'street', 'number','elevation']]
locations = locations.set_index('id', drop = False)
airly_data_2018 = pd.read_csv("data/data.csv")
```

## Correlations between pm10 and humidity, temperature and pressure 

```python
start_date = "2018-01-31T11:00:00"
end_date = "2018-12-31T23:00:00"
for loc in locations['id']:
    try:
        data = extract_data(airly_data_2018, loc, start_date, end_date , ['pm10', 'humidity', 'temperature', 'pressure'])
        print(locations[locations['id']==loc])
        scatter_matrix(data)
        plt.show()
    except KeyError:
        pass
```
